```python
# ruff: noqa: F401
from typing import Dict, List, Any
from supergene.mongo import Mongo
from supergene.v3 import V3
from rich.pretty import Pretty
from rich.table import Table
from rich.console import Console
from rich.traceback import install as tr_install
from os import getenv
from pymongo import MongoClient
from bunnet import init_bunnet
from dotenv import load_dotenv

load_dotenv()

def connect() -> None:
    """Connect to the MongoDB database."""
    uri = getenv("SUPERGENE", "op://Personal/ixzlwkey4nyc2w54fathyi4ilq/Database/uri")
    client: MongoClient = MongoClient(uri)
    docs: List[Any] = [V3]
    init_bunnet(
        database=client.supergene,
        document_models=docs,
    )

console = Console()
tr_install(console=console)
```

```python


def get_totals() -> Dict[int, int]:
    """Get the number of chapters in each book.
    
    Args:
        mongo: A Mongo instance.
    """
    connect()
    console=Console()
    totals = {}
    for book in range(1,11):
        console.print(f"Book {book}")
        chapters = V3.find(V3.book == book).to_list()
        chapter_count = len(chapters)
        console.print(f"Chapter Count: {chapter_count}")
        totals[book] = chapter_count
    console.print(Pretty(totals))
    return totals
```
